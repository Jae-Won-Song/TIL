```python
import matplotlib.pyplot as plt
import numpy as np

matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 그래프에서의 한글 사용
matplotlib.rcParams['axes.unicode_minus'] = False


plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
plt.xlabel('X-Axis', labelpad = 15, loc = 'right')
plt.ylabel('Y-Axis', labelpad = 20, loc = 'top')
plt.show()
```
